##**Polars**



Считайте датасет из файла train.csv (это данные о выживаемости на Титанике) с помощью polars

```python
import polars as pl

df = pl.read_csv('train.csv')
```

Выведите основную информацию о датасете: информацию о типах данных, число пропусков, средние значения и т.д.

```python
df.describe()
```

Посчитайте количество пассажиров каждого класса (Pclass)

```python
class_counts = df.groupby('Pclass').agg(pl.col('Pclass').count().alias('Count'))
class_counts
```

Выведите количество выживших мужчин и женщин на корабле

```python
survived_gender_counts = df.filter(df['Survived'] == 1).groupby('Sex').agg(pl.col('Sex').count().alias('Count'))
survived_gender_counts
```

Выведите часть таблицы с пассажирами, возраст которых больше 44 лет

```python
older_than_44 = df.filter(df['Age'] > 44)
older_than_44
```

##**Ускорение работы с pandas**



Считайте датасет из файла train.csv (это данные о выживаемости на Титанике) с помощью pandas

```python
import pandas as pd

df = pd.read_csv('train.csv')
```

Посчитайте средний возраст пассажиров и его стандартное отклонение с помощью bottleneck

```python
pip install bottleneck
```

```python
import bottleneck as bn

mean_age = bn.nanmean(df['Age'])
print("Средний возраст пассажиров:", mean_age)

std_age = bn.nanstd(df['Age'])
print("Стандартное отклонение возраста пассажиров:", std_age)
```

Для каждого пассажира умножьте значение столбца Fare на 1.3, если класс его билета - 1 или 2, и на 1.1, если класс его билета - 3. Сохраните результаты как новый столбец Fare_new. (При реализации расчета используйте любые альтернативы методу iterrows)

```python
df['Fare_new'] = df.apply(lambda row: row['Fare'] * 1.3 if row['Pclass'] in [1, 2] else row['Fare'] * 1.1, axis=1)
df[['Pclass', 'Fare', 'Fare_new']]
```

*Данные в колонках bedrooms, bathrooms, stories и parking лучше сделать типом int8, area типом int16, а price типом int32, тип int64 для них избыточен*

Поменяйте типы данных столбцов датафрейма на выбранные вами в прошлом пункте и сравните потребление памяти до и после оптимизации

```python
import numpy as np
df['price'] = df['price'].astype(np.int32)
df['area'] = df['area'].astype(np.int16)
df['bedrooms'] = df['bedrooms'].astype(np.int8)
df['bathrooms'] = df['bathrooms'].astype(np.int8)
df['stories'] = df['stories'].astype(np.int8)
df['parking'] = df['parking'].astype(np.int8)
```

*До изменения типа данных датасет весил 55.5 Kб, теперь весит 35,3 Кб*
